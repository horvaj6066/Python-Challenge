````python
'''## PyPoll

![Vote-Counting](Images/Vote_counting.jpg)

* In this challenge, you are tasked with helping a small, rural town modernize its vote-counting process. (Up until now, Uncle Cleetus had been trustfully tallying them one-by-one, but unfortunately, his concentration isn't what it used to be.)

* You will be give a set of poll data called [election_data.csv](PyPoll/Resources/election_data.csv). The dataset is composed of three columns: `Voter ID`, `County`, and `Candidate`. Your task is to create a Python script that analyzes the votes and calculates each of the following:

  * The total number of votes cast

  * A complete list of candidates who received votes

  * The percentage of votes each candidate won

  * The total number of votes each candidate won

  * The winner of the election based on popular vote.

* As an example, your analysis should look similar to the one below:

  ```text
  Election Results
  -------------------------
  Total Votes: 3521001
  -------------------------
  Khan: 63.000% (2218231)
  Correy: 20.000% (704200)
  Li: 14.000% (492940)
  O'Tooley: 3.000% (105630)
  -------------------------
  Winner: Khan
'''

#first import needed libraries 
import os 
import csv
import re

#initialize variables
voter_ID=[]
county=[]
candidate=[]
unique_candidates=[]
vote_count=[]

#set path for file
csvpath = os.path.join("election_data.csv")

#open the CSV
with open(csvpath, newline="", encoding='utf-8') as csvfile: 
    csvreader = csv.reader(csvfile)
    
    for row in csvreader:  #loop to read the csv file into 3 lists
        voter_ID.append(row[0])  #append first filed to voter_ID
        county.append(row[1])   #also store the profit in a temp variable 
        candidate.append(row[2]) #add canddiate 

#this loop is to separate the all candidates into a new list of unique_candidates        
for i in candidate:
    if i in unique_candidates:
        continue
    else:
        unique_candidates.append(i)
    
for j in range(1,len(unique_candidates)):  # this loop tallies up each candidates votes and prints 

    print("\n"+ unique_candidates[j],end=" ")  #print unique candidate names 
    temp_votes = candidate.count(unique_candidates[j]) #tally of votes/candidate
    vote_count.append(temp_votes)  #add votes/candidate to the vote count list 
    print(temp_votes, end=" ")
    print("votes, which is", end=" ") 
    percent_votes=temp_votes/len(candidate)*100 
    print('{0:3.0f}'.format(percent_votes), end=" ")  # 'format so that it rounds properly
    print("percent of total votes.") 
    
winner = vote_count.index(max(vote_count))  #winner points to the candidate w most votes  
del unique_candidates[0]  #still have header in there at position zero so deleting 
print("\n"+"Winner: ",unique_candidates[winner])  #and the winner is....

#lets also output to txt file
with open("PyPollOUT.txt", 'w+') as the_file:   #not using csv because don't want comma delimited
    #next few lines writing are doing the same as above prints but to file
    the_file.write("Winner: " + unique_candidates[winner]+"\r")  #The winner is....
       
    for j in range(0, len(unique_candidates)): # this loop goes thru the unique candidates and writes their info to file
        votes = vote_count[j] #temp variable to convert the vote count to string
        the_file.write(unique_candidates[j] + " ") #candidate name
        the_file.write(str(votes)+ " votes, which is")  #total votes for this candidate 
        the_file.write(str('{0:3.0f}'.format(votes/len(candidate)*100)) +" percent of total votes") # % of total
        the_file.write("\r")

#done
````
