```python
'''
***********PyBank*************************************************************
* In this challenge, you are tasked with creating a Python script for analyzing the financial records of your company. You will give a set of financial data called [budget_data.csv](PyBank/Resources/budget_data.csv). The dataset is composed of two columns: `Date` and `Profit/Losses`. (Thankfully, your company has rather lax standards for accounting so the records are simple.)
*
* Your task is to create a Python script that analyzes the records to calculate each of the following:
* The total number of months included in the dataset
* The net total amount of "Profit/Losses" over the entire period
* The average of the changes in "Profit/Losses" over the entire period
* The greatest increase in profits (date and amount) over the entire period
* The greatest decrease in losses (date and amount) over the entire period
* As an example, your analysis should look similar to the one below:
********************************************************************************
  text
  Financial Analysis
  ----------------------------
  Total Months: 86
  Total: $38382578
  Average  Change: $-2315.12
  Greatest Increase in Profits: Feb-2012 ($1926159)
  Greatest Decrease in Profits: Sep-2013 ($-2196167)
  '''

#first import needed libraries 
import os 
import csv
import re

#lets declare our lists and variables 
date=[]
profit=[]
chg_in_profit=[]
total_profit=0
temp_profit=0
#need the months in here to match a number to a 3 letter abbreviation for tallying the total months
month=['Jan', 'Feb', 'Mar', 'Apr','May','Jun','Jul','Aug','Sep','oct','Nov','Dec']


#set path for file
csvpath = os.path.join(".", "C:\Users\Jon P Horvath\py\Python-Challenge\PyBank", "budget_data.csv")

#open the CSV
with open(csvpath, newline="", encoding='utf-8') as csvfile: 
    csvreader = csv.reader(csvfile)
  
    for row in csvreader:  #loop to read the csv file
        date.append(row[0])  #append first filed to date
        temp_profit=row[1]   #also store the profit in a temp variable 
        profit.append(row[1]) #append 2nd filed to profit 

 
count=len(date) #make count represent how many elements are in the list of dates 
profit[0]=0 #this is to make the header a number that won't mess with our calcs

# a loop to convert profit from string to integer and also create a new list chg_in_profit 
for i in range(1, len(profit)):   # a loop to convert profit from string to integer 
    profit[i] = int(profit[i]) #not sure why i can't do this when I'm doing the append
    if i > 1: 
        chg_in_profit.append(profit[i]-profit[i-1]) #this creates a list that contains m/m chge in profit  
    
total_profit = sum(profit) #summarize total profit over the period 
maxpos = chg_in_profit.index(max(chg_in_profit))+1 #maxpos is the position of the max m/m chge in profits
minpos = chg_in_profit.index(min(chg_in_profit))+1 #minpos is the position of the max m/m chge in profits
mean = int(sum(chg_in_profit)/len(chg_in_profit)) #mean is the avg m/m chge in profits


#need to split the date in order to calculate how many months in total period
first_month,first_year = date[1].split('-') #split the first date into two variables 
last_month, last_year= date[count-1].split('-') #split last date into two variables 

#loop below is to convert months to numbers to help calculate length of the overall period in months
for j in range(0,11):
    if first_month==month[j]:
        first_month=j+1
    if last_month==month[j]:
        last_month=j+1

#calculate how many months from beginning to end 
months= (int(last_year)-int(first_year))*12 + (last_month-first_month+1) 

#print out all the deliverables to the screen 
print("This is based on " + str(months) + " months of data")
print("Total profit over this period is $"+str(total_profit))
print("The average month/month change in profits was: $", str(mean))
print("Greatest month/month increase in profits was: $" + str(max(chg_in_profit)) + " during " + date[maxpos])  
print("Greatest month/month decrease in profits was: $" + str(min(chg_in_profit)) + " during " + date[minpos])

#also do the output to all the deliverables to file as per directions
with open("PyBank.txt", 'w+') as the_file:   #not using csv writer because don't want comma delimited
    #next few lines writing exactly the same as above prints
    the_file.write("This is based on " + str(months) + " months of data")
    the_file.write("\rTotal profit over this period is $") 
    the_file.write(str(total_profit)) 
    the_file.write("\rThe average month/month change in profits was: $")
    the_file.write(str(mean))
    the_file.write("\rGreatest month/month increase in profits was: $" + str(max(chg_in_profit)) + " during " + date[maxpos])  
    the_file.write("\rGreatest month/month decrease in profits was: $" + str(min(chg_in_profit)) + " during " + date[minpos])



```
